```python
#first we need 3 classes, Card, Deck and player

import random
suits = ('Diamonds','Hearts','Clubs','Spades')
numbers = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
points = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':11}

#secondly, we add in the finances:
#thinking of a blackjack(amt) where amt is the betting amt. 
#another sub class for bank account? Wiht withdrawl and deposits.
#add in special conditions where BLACKJACK doubles the winners money, no 5 card nonsense
#Casino account database

database = {}

def newaccount():
    newacc = input('Welcome to the 7luck Casino!! Please enter new account name:')
    database.update({newacc : 100})
    print(f'Welcome {newacc}, all new players receive complementary $100. GOOD LUCK!!')
    
#Thirdly,we need a game on where betting amount can be controlled each time

def game_on():
    ans = ''
    while ans not in ['y','n']:
        ans = input("Would you like to play again? y/n :")
        if ans not in ['y','n']:
            print('Sorry your input is incorrec!!')
    return ans == 'y'

#lastly, clean this up by clearing outputs through (from IPython.display import clear_output)
#and use clear_output() to tidy things up.

from IPython.display import clear_output
```

```python
class Card:
    def __init__(self,suit,number):
        self.suit = suit
        self.number = number
        self.point = points[number]
        
    def __str__(self):
        return f'{self.number} of {self.suit}'
        
class Deck:
    def __init__(self):
        self.allcards = []
        for suit in suits:
            for number in numbers:
                self.allcards.append(Card(suit,number))
                
    def shuffle(self):
        random.shuffle(self.allcards)
        
    def deal_one(self):
        return self.allcards.pop()
    
class Player():
    def __init__(self,name,balance = 0):
        self.name = name
        self.balance = balance
        self.hand = []
        self.cardlist = []
        
    def add_cards(self,newcard):
        self.hand.append(newcard)
        self.cardlist.append(str(newcard))
        #or self.cardlist.append(newcard.__str__())
        self.cards = ' & '.join(self.cardlist)
        
    def value(self):
        self.score=0
        for card in self.hand:
            self.score += card.point 
        return self.score
        
    def __str__(self):
        return f'{self.name} has {self.cards}'  
```

```python
def blackjack():
    #Player account
    temp = ''
    keys = database.keys()
    while temp not in keys:
        clear_output()
        temp = input('Please enter account name (case-sensitive):')
        while temp not in keys:
            ans = input('Sorry account name is unregistered. Would you like to register a new account? y/n :')
            if ans == 'y':
                clear_output()
                newaccount()
            elif ans == 'n':
                clear_output()
                break
            else:
                clear_output()
                print('Incorrect input!!')
    
    #Preparing the player & dealer
    dealer = Player('Dealer')
    player = Player(temp, database[temp])
    newbalance = player.balance
    
    gameon = True
    while gameon:
        
        #setting the table: resetting the deck
        newdeck = Deck()
        newdeck.shuffle()
        
        #resetting & preparing the cards
        player.hand = []
        dealer.hand = []
        player.cardlist = []
        dealer.cardlist = []
        dealer.add_cards(newdeck.deal_one())
        player.add_cards(newdeck.deal_one())
        dealer.add_cards(newdeck.deal_one())
        player.add_cards(newdeck.deal_one())

        player.score = player.value()
        dealer.score = dealer.value()
        
        bet = -1
        if newbalance == 0:
            print(f"{player.name}'s account balance is ${newbalance}.\n{player.name.upper()} FILLED FOR BANKRUPTCY")
            return 'Goodbye'
            
        while bet not in range(newbalance+1):
            print(f'Account balance: $ {newbalance}')
            bet = input('Enter bet amount: $')
            if bet.isdigit() == True:
                bet = int(bet)
                if bet > newbalance:
                    clear_output()
                    print('Insufficient balance!!')
                else:
                    pass
            else:
                clear_output()
                print('Incorrect input!!')
        '''
        Also can:
        while bet not in range(newbalance + 1):
            print(f'Account balance: $ {newbalance}')
            try:
                bet = int(input('Enter bet amount: $'))
            except:
                clear_output()
                print('Incorrect input!!')
            else:
                if bet > newbalance:
                    clear_output()
                    print('Insufficient balance!!')
                else:
                    pass
        '''
        

        print(f'Welcome {player.name} to BlackJack!!')
        
        #Insert blackjack winning conditions/special conditions here
        if player.score == 21 or player.score == 22:
            print(f'BLACKJACK!! Your cards are {player.cards} and you WIN!!')
            newbalance += 2*bet
            print (f'New account balance: $ {newbalance}')
            database[player.name] = newbalance
            gameon = game_on()
            clear_output()
            
        elif dealer.score == 21 or dealer.score == 22:
            print(f"BLACKJACK!! Dealer's cards are {dealer.cards} and he WINS!!")
            newbalance -= 2*bet
            print (f'New account balance: $ {newbalance}')
            database[player.name] = newbalance
            gameon = game_on()
            clear_output()
        
        #game continues as usual
        else:
            print(f'Your cards are {player.cards}\nYour score is {player.score}')
            print(f'Dealer currently has {dealer.cardlist[0]}')


            #player turn to hit or stand
            hit = True
            while hit == True:            
                decision = ''
                acceptable_values = ['hit','stand']
                while decision not in acceptable_values:
                    decision = input('Would you like to hit or stand?:')
                    if decision not in acceptable_values:
                        print('Sorry your input is incorrect')
                        
                #if decision is to hit
                if decision == 'hit':
                    player.add_cards(newdeck.deal_one())
                    player.score = player.value() 
                    if player.score <= 21:
                        print(f"Your cards are {player.cards}\nYour score is {player.score}")

                    #above 21
                    else:
                        for card in player.hand:

                            #above 21 but ace
                            if card.point == 11:
                                player.score = player.score - 10
                                print(f"Your cards are {player.cards}\nYour score is {player.score}")
                                if player.score > 21:
                                    print (f'{player.name} BUSTS!! Dealer wins')
                                    newbalance -=bet
                                    print (f'New account balance: $ {newbalance}')
                                    database[player.name] = newbalance
                                    gameon = game_on()
                                    clear_output()
                                    break
                                else:
                                    break

                        #above 21 but no ace
                        else:
                            print(f"Your cards are {player.cards}\nYour score is {player.score}")
                            print (f'{player.name} BUSTS!! Dealer wins')
                            newbalance -= bet
                            print (f'New account balance: $ {newbalance}')
                            database[player.name] = newbalance
                            gameon = game_on()
                            clear_output()
                            break

                #if decision is to stand
                else:
                    hit = False

            #dealers turn
            while hit == False:
                while dealer.score <= 21 :

                    #dealer has less than player and less than 21 
                    if dealer.score < player.score or dealer.score == player.score:
                        dealer.add_cards(newdeck.deal_one())
                        dealer.score = dealer.value()

                    #dealer has more than player and less than 21    
                    else:
                        print (f"Dealer has {dealer.cards}. Dealer's {dealer.score} is more than your {player.score}, Dealer wins!!")
                        newbalance -= bet
                        print (f'New account balance: $ {newbalance}')
                        database[player.name] = newbalance
                        gameon = game_on()
                        clear_output()
                        hit = None
                        break

                #above 21
                else:
                    for card in dealer.hand:

                        #above 21 but ace
                        if card.point == 11:
                            dealer.score = dealer.score - 10
                            print(f"Dealer's cards are {dealer.cards}\nHis score is {dealer.score}")

                            #dealer busts with ace
                            if dealer.score > 21:
                                print (f'Dealer BUSTS!! {player.name} wins')
                                newbalance += bet
                                print (f'New account balance: $ {newbalance}')
                                database[player.name] = newbalance
                                gameon = game_on()
                                clear_output()
                                hit = None
                                break

                            #dealer wins with ace
                            elif dealer.score > player.score:
                                print (f"Dealer has {dealer.cards}. Dealer's {dealer.score} is more than your {player.score}, Dealer wins!!")
                                newbalance -= bet
                                print (f'New account balance: $ {newbalance}')
                                database[player.name] = newbalance
                                gameon = game_on()
                                clear_output()
                                hit = None
                                break

                            #dealer hasn't bust and hasn't won the player
                            else:
                                break

                    #above 21 but no ace
                    else:
                        print(f"Dealer's cards are {dealer.cards}\nHis score is {dealer.score}")
                        print (f'Dealer BUSTS!! {player.name} wins')
                        newbalance += bet
                        print (f'New account balance: $ {newbalance}')
                        database[player.name] = newbalance
                        gameon = game_on()
                        clear_output()
                        hit = None
                        break

    clear_output()
    print('Thanks for playing!')
    return f"{player.name}'s final account balance is ${newbalance}"
        
```
